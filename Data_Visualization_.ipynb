```python
import pandas as pd
import numpy as np
```

```python
# A matplotlib workflow

# 0. Import and get matplotlib ready
import matplotlib.pyplot as plt

# 1. Prepare data
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]

# 2. Setup plot
fig, ax = plt.subplots(figsize=(10,10))

# 3. Plot data
ax.plot(x, y)

# 4. Customize plot
ax.set(title="Sample Simple Plot", xlabel="x-axis", ylabel="y-axis")

# 5. Save & show
fig.savefig("Test.png")
```

```python
cars_prices = {"Toyota": 500000,
               "BMW": 1000000,
               "Apha Romeo": 2800000}
fig,ax = plt.subplots()
ax.bar(cars_prices.keys(),cars_prices.values())
ax.set(title="Shazly store",
       xlabel = 'Car Type',
       ylabel = "price")
```

```python
cars_prices = {"Toyota": 500000,
               "BMW": 1000000,
               "Apha Romeo": 2800000}
fig,ax = plt.subplots()
ax.barh(list(cars_prices.keys()),list(cars_prices.values()))
ax.set(title="Shazly store",
       ylabel = 'Car Type',
       xlabel = "price")
```

```python
x = np.random.randn(1000)
fig,ax = plt.subplots()
ax.hist(x)
```

```python
# Option 1: Create multiple subplots
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2,
                                             ncols=2,
                                             figsize=(10, 5))
# Plot data to each axis
ax1.plot(x, x/2);
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(cars_prices.keys(), cars_prices.values());
ax4.hist(np.random.randn(1000));
```

```python
# Option 2: Create multiple subplots
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(10, 5))

# Index to plot data
ax[0, 0].plot(x, x/2);
ax[0, 1].scatter(np.random.random(10), np.random.random(10));
ax[1, 0].bar(cars_prices.keys(), cars_prices.values());
ax[1, 1].hist(np.random.randn(1000));
```

```python
df["Price"] = df["Price"].astype(int)
```

```python
df["Total sales"] = df["Price"].cumsum()
```
